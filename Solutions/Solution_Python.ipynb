```python
import matplotlib.pyplot as plt

# Hard-coded data for markets and sales
markets = ['Africa', 'Asia Pacific', 'Europe', 'LATAM', 'USCA']
sales = [783773, 4042658, 3287336, 2164605, 2364129]

# Hard-coded reference line value
reference_value = 2000000

# Create a column chart
plt.figure(figsize=(10, 6))
plt.bar(markets, sales)

# Add labels and title
plt.xlabel('Market')
plt.ylabel('Sales (Whole Numbers)')
plt.title(f'Total Sales by Market with Reference Line at {reference_value}')

# Add a reference line at the user-specified value
plt.axhline(y=reference_value, color='r', linestyle='--', label=f'Reference Line: {reference_value}')

# Add legend and display the chart
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Hard-coded data for markets and sales
markets = ['Africa', 'Asia Pacific', 'Europe', 'LATAM', 'USCA']
sales = [783773, 4042658, 3287336, 2164605, 2364129]

# Ask the user for the reference line value
reference_value = int(input("Enter the reference line value: "))

# Determine colors based on whether the sales are above or below the reference value
colors = ['blue' if sale > reference_value else 'gray' for sale in sales]

# Create a column chart
plt.figure(figsize=(10, 6))
plt.bar(markets, sales, color=colors)

# Add labels and title
plt.xlabel('Market')
plt.ylabel('Sales (Whole Numbers)')
plt.title(f'Total Sales by Market with Reference Line at {reference_value}')

# Add a reference line at the user-specified value
plt.axhline(y=reference_value, color='r', linestyle='--', label=f'Reference Line: {reference_value}')

# Add legend and display the chart
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
